Importing the dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Importing the Boston House Price Dataset

```python
house_price_dataset = sklearn.datasets.fetch_california_housing()
```

```python
print(house_price_dataset)
```

```python
# loading the data set to a pandas dataframe(to make data more structured)
house_price_dataframe = pd.DataFrame(house_price_dataset.data,columns = house_price_dataset.feature_names)
#column section mentioned to come under proper feature with name if not then mentions 0,1,2..
```

```python
#print First 5 rows of our dataFrame
house_price_dataframe.head()
```

```python
# add the target (price) column to the DataFrame
house_price_dataframe['price'] = house_price_dataset.target
```

```python
#print First 5 rows of our dataFrame
house_price_dataframe.head()
```

```python
# checking the no of rows and columns in the dataframe
house_price_dataframe.shape
```

```python
#check for missing values(isnull fn checks yhe no of missing values)
house_price_dataframe.isnull().sum()
```

```python
#statistical measures of the dataset - use describe fn
house_price_dataframe.describe()
```

Understanding the corelation between various features in the dataset

```python
correlation = house_price_dataframe.corr()
```

```python
#construct a heatmap to understand the correlation
#fmt for no. of fp after decimal,annot_kws for size of text
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
# white colour - ve correlated(one inc other dec)
```

splitting the data and label(target) ie price

```python
# drop a col : axis=1,row : axis=0
X = house_price_dataframe.drop(['price'],axis = 1)
Y = house_price_dataframe['price']
```

```python
print(X)
print(Y)
```

splitting the data into training and test data

```python
#let 80 percent for train and 20 per for test
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model training

XGBoost Regresor


```python
#loading the model
model = XGBRegressor()
```

```python
# training the model with X_train
model.fit(X_train, Y_train)
```

Evaluation

Prediction on training data

```python
#accuracy score cant be used in regression(only for classification) as evaluation metric and we use r squared error and mean error
#accuracy for prediction on training data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# r squared error (r2 finds variance b/w Y and train..)
score_1 = metrics.r2_score(Y_train,training_data_prediction)
print("R squared error : ",score_1)
# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_train,training_data_prediction)
print("Mean Absolute Error : ",score_2)
```

Visualizing the actual Prices and predicted prices

```python
plt.scatter(Y_train,training_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices vs Predicted Prices")
plt.show()
```

Prediction on test data

```python
#accuracy for prediction on training data
test_data_prediction = model.predict(X_test)
```

```python
# r squared error (r2 finds variance b/w Y and train..)
score_1 = metrics.r2_score(Y_test,test_data_prediction)
print("R squared error : ",score_1)
# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_test,test_data_prediction)
print("Mean Absolute Error : ",score_2)
```
